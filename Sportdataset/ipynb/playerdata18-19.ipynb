```python
#passing
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

# Set the URL of the website you want to scrape
url = "https://fbref.com/en/comps/9/2018-2019/stats/2018-2019-Premier-League-Stats"

# Set the path to your Chrome WebDriver executable
service = Service(executable_path="/Users/morgan/Desktop/chromedriver-mac-arm64/chromedriver")
# Create a Chrome WebDriver instance
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service,options=options)

# Navigate to the URL
driver.get(url)

# Wait for the page to load (you may need to adjust the wait time)
driver.implicitly_wait(10)

# Find the table containing player stats
stats_table = driver.find_element(By.ID, "stats_standard")

# Find all rows in the table
rows = stats_table.find_elements(By.TAG_NAME, "tr")

# Initialize lists to store data
player_names = []
MP = []
Gls = []
AST = []
CRDY = []
CRDR = []



# Loop through the rows and extract data
for row in rows[1:]:  # Skip the header row
    columns = row.find_elements(By.TAG_NAME, "td")
    if len(columns) >= 18:
        player_name = columns[0].text
        Match = columns[6].text
        goals = columns[10].text
        assist = columns[11].text
        yellow = columns[16].text
        red = columns[17].text


        player_names.append(player_name)
        MP.append(Match)
        Gls.append(goals)
        AST.append(assist)
        CRDY.append(yellow)
        CRDR.append(red)


# Close the Chrome WebDriver
driver.quit()

# Print or process the collected data as needed
for i in range(len(player_names)):
    print(f"Player Name: {player_names[i]}, mp: {MP[i]}, Gls: {Gls[i]}, Ast: {AST[i]}, CRDY: {CRDY[i]}, CRDR: {CRDR[i]}")
```
